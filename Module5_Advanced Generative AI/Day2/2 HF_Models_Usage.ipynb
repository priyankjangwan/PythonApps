Vision Transformer for Image Classification

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
```

```python
from PIL import Image as img
# PIL (Python Imaging Library) to work with images
# Image is renamed to img using as img just to avoid confusion or naming conflicts.
# This lets you open, display, crop, resize, or transform images in Python.

from IPython.display import Image, display
```

```python
FILE_NAME = "/content/ball.jpeg"
FILE_NAME = "/content/ws.jpg"
```

```python
display(Image(FILE_NAME, width = 700, height = 400))
```

```python
# image_array = img.open('ball.jpeg')
# type(image_array)

image_array = img.open('ws.jpg')
type(image_array)
```

```python
# https://arxiv.org/pdf/2010.11929

feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
# Extract features

inputs = feature_extractor(images = image_array,
                           return_tensors="pt")

outputs = model(**inputs)
logits = outputs.logits
```

```python
# logits
logits.shape
```

```python
# logits
```

```python
predicted_class_idx = logits.argmax(-1).item()
```

```python
print(predicted_class_idx)
# https://deeplearning.cms.waikato.ac.nz/user-guide/class-maps/IMAGENET/
```

```python
print("Predicted class:", model.config.id2label[predicted_class_idx])
```

## Stable Diffusion

```python
!pip install diffusers -q
# The diffusers library is developed by Hugging Face and is commonly
# used for working with diffusion models.
# Diffusion models are a type of machine learning model often used
# for generating images, audio, and other types of data.
```

```python
!pip install accelerate -q
# The accelerate library is developed by Hugging Face
# and is designed to simplify the process of training and running machine
# learning models, especially deep learning models, with multiple GPUs or TPUs.
# It helps manage resources efficiently, distribute training across devices, and
#  optimize performance.
```

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda") # this moves the entire pipeline(model weights and tensors) to GPU memory.
# If you skip this, the model will run on CPU — very slow for image generation.
```

```python
prompt = "a photo of an astronaut riding a horse on mars planet"
image = pipe(prompt).images[0]

image.save("astronaut_rides_horse.png")
```
